```python
import pandas as pd
dict1 ={'customer_id':[1,2,3,4,5],'Product':['Oven','oven','television','oven','oven']}
df1 = pd.DataFrame(dict1)
df1
```

```python
import pandas as pd
dict1 ={'customer_id':[2,4,6,7,9],'State':['melbourne','sydney','perth','The Gabba','india']}
df2 = pd.DataFrame(dict1)
df2
```

```python
inner_join_df = pd.merge(df1,df2,on='customer_id',how='inner')
outer_join_df = pd.merge(df1,df2,on='customer_id',how='outer')
inner_join_df
outer_join_df
```

```python
left_join_df = pd.merge(df1,df2,on='customer_id',how='left')
#right_join_df = pd.merge(df1,df2,on='customer_id',how='right')
left_join_df
#right_join_df
```

```python
upper_join_df = df1.join(df2,lsuffix='_left',rsuffix='_right')
upper_join_df
lower_join_df = df2.join(df1,lsuffix='_left',rsuffix='_right')
lower_join_df
```
