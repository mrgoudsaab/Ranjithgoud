```python
file_path = '/content/Movie.csv'
Movie = pd.read_csv(file_path)
print(Movie.head())
```

```python
len(Movie.movie.unique())
```

```python
len(Movie.userId.unique())
```

```python
user_Movie_rating = Movie.pivot(index='userId',
                                 columns='movie',
                                 values='rating')
print(user_Movie_rating.head())
```

```python
user_Movie_rating.fillna(0, inplace=True)
user_Movie_rating.head()
```

```python
user_Movie_rating
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine = cosine_similarity(user_Movie_rating)
cosine_df = pd.DataFrame(cosine, index=user_Movie_rating.index, columns=user_Movie_rating.index)
print(cosine_df.head())
```

```python
from scipy.spatial.distance import correlation

vec1 = [5,4,0,0]
vec2 = [4,5,3,0]
correlation_distance = correlation(vec1, vec2)
print(correlation_distance)
```

```python
user_sim = 1 - pairwise_distances(user_Movie_rating, metric='cosine')

```
